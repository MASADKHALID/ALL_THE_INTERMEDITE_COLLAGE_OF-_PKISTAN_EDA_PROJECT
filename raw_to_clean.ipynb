```python
import pandas as pd #IMPORT LIBRARY
```

```python
data_frame=pd.read_csv("All the Intermediate Colleges in Pakistan.csv",index_col=False) #REEAD DATASET
```

```python
data_frame #SHOW DATA FRAME
```

```python
data_frame.columns #DISPLAY COLUMNS NAMES
```

```python
data_frame = data_frame.rename(columns={'Study Program': 'Study_Program'}) # RENAME STUDY PROGRAME 
```

```python
data_frame # SHOW DATA FRAME
```

```python
data_frame.dtypes # DISPLAY DATATYPES
```

```python
# MAINUPLATE COLUMN RATING IN DATAFRAME
data_frame['Rating'].unique() # DISPLAY UNIQUE VALUES OF RATING
```

```python
data_frame['Rating'] = data_frame['Rating'].str.replace('*', '', regex=False).astype(float) # REPELACE  *  AND CONVERT DATTYPE
```

```python
data_frame['Rating'].unique() # CHECHK VALES
```

```python
data_frame.dtypes #CHECK DATATYPE
```

```python
# MAINUPLATE COLUMN AFFILIATION IN DATAFRAME
data_frame['Affiliation'].unique() # DISPLAY UNIQUE VALUES OF AFFILIATION
```

```python
data_frame['Affiliation']=data_frame['Affiliation'].replace('nan', None) # REPLACE NAH VALUE
```

```python
data_frame['Affiliation'].unique() # CHECK
```

```python
# MAINUPLATE COLUMN SECTOR IN DATAFRAME
data_frame['Sector'].unique() #DISPLAY UNIQUE VALUES OF AFFILIATION
```

```python
data_frame['Sector']=data_frame['Sector'].replace('nan', None) # REPLACE NAH VALUE
```

```python
data_frame['Sector'].unique() # check values
```

```python
data_frame['ID'] = range(1, len(data_frame) + 1) #add new column generate id for each record
```

```python
data_frame # DISPLPY DATAFRAME 
```

```python
data_frame.insert(0, 'ID', data_frame.pop('ID')) # ADD AS 1 COLUMN
```
